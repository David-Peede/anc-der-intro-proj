```python
def load_population_results():
    """
    ###########################################################################
    INPUT: N/A
    ---------------------------------------------------------------------------
    OUTPUT: Genome-wide site pattern and introgression metrics.
    ###########################################################################
    """
    # Load all of the site pattern results.
    genome_df = pd.read_csv(
        './tgp_altai_site_patterns.csv',
        names=['POP', 'ABBA', 'BABA', 'BAAA', 'ABAA', 'ABBA_HOM', 'BABA_HOM', 'BAAA_HOM', 'ABAA_HOM'],
    )
    # Calculate site pattern differences.
    genome_df['ABBA-BABA'] = genome_df['ABBA'] - genome_df['BABA']
    genome_df['BAAA-ABAA'] = genome_df['BAAA'] - genome_df['ABAA']
    # Calculate introgression detection metrics.
    genome_df['D'] = (genome_df['ABBA'] - genome_df['BABA']) / (genome_df['ABBA'] + genome_df['BABA'])
    genome_df['Danc'] = (genome_df['BAAA'] - genome_df['ABAA']) / (genome_df['BAAA'] + genome_df['ABAA'])
    genome_df['D+'] = ((genome_df['ABBA'] - genome_df['BABA']) + (genome_df['BAAA'] - genome_df['ABAA']))\
        /((genome_df['ABBA'] + genome_df['BABA']) + (genome_df['BAAA'] + genome_df['ABAA']))
    # Calculate introgression quantification metrics.
    genome_df['fhom'] = (genome_df['ABBA'] - genome_df['BABA']) / (genome_df['ABBA_HOM'] - genome_df['BABA_HOM'])
    genome_df['fanc'] = (genome_df['BAAA'] - genome_df['ABAA']) / (genome_df['BAAA_HOM'] - genome_df['ABAA_HOM'])
    genome_df['f+'] = ((genome_df['ABBA'] - genome_df['BABA']) + (genome_df['BAAA'] - genome_df['ABAA']))\
        /((genome_df['ABBA_HOM'] - genome_df['BABA_HOM']) + (genome_df['BAAA_HOM'] - genome_df['ABAA_HOM']))
    # Drop the unecessary columns.
    genome_df = genome_df.drop(columns=['ABBA_HOM', 'BABA_HOM', 'BAAA_HOM', 'ABAA_HOM'])
    # Intialize the population order.
    tgp_pop_array = np.array([
        'CEU', 'FIN', 'GBR', 'IBS', 'TSI',
        'CHB', 'CHS', 'CDX', 'JPT', 'KHV',
        'BEB', 'GIH', 'ITU', 'PJL', 'STU',
        'CLM', 'MXL', 'PEL', 'PUR',
    ])
    # Make the population column categorical.
    genome_df['POP'] = pd.Categorical(
        genome_df['POP'],
        categories=tgp_pop_array,
        ordered=True,
    )
    # Sort by the population column.
    genome_df.sort_values(['POP'], inplace=True, ignore_index=True)
    return genome_df
```

```python
def load_bootstrapped_results():
    """
    ###########################################################################
    INPUT: N/A
    ---------------------------------------------------------------------------
    OUTPUT: Genome-wide site pattern and introgression metrics bootstraps.
    ###########################################################################
    """
    # Load all of the site pattern results.
    bootstrap_df = pd.read_csv(
        './tgp_altai_bootstrapped_site_patterns.csv',
        names=['POP', 'ABBA', 'BABA', 'BAAA', 'ABAA', 'ABBA_HOM', 'BABA_HOM', 'BAAA_HOM', 'ABAA_HOM'],
    )
    # Calculate introgression detection metrics.
    bootstrap_df['D'] = (bootstrap_df['ABBA'] - bootstrap_df['BABA']) / (bootstrap_df['ABBA'] + bootstrap_df['BABA'])
    bootstrap_df['Danc'] = (bootstrap_df['BAAA'] - bootstrap_df['ABAA']) / (bootstrap_df['BAAA'] + bootstrap_df['ABAA'])
    bootstrap_df['D+'] = ((bootstrap_df['ABBA'] - bootstrap_df['BABA']) + (bootstrap_df['BAAA'] - bootstrap_df['ABAA']))\
        /((bootstrap_df['ABBA'] + bootstrap_df['BABA']) + (bootstrap_df['BAAA'] + bootstrap_df['ABAA']))
    # Calculate introgression quantification metrics.
    bootstrap_df['fhom'] = (bootstrap_df['ABBA'] - bootstrap_df['BABA']) / (bootstrap_df['ABBA_HOM'] - bootstrap_df['BABA_HOM'])
    bootstrap_df['fanc'] = (bootstrap_df['BAAA'] - bootstrap_df['ABAA']) / (bootstrap_df['BAAA_HOM'] - bootstrap_df['ABAA_HOM'])
    bootstrap_df['f+'] = ((bootstrap_df['ABBA'] - bootstrap_df['BABA']) + (bootstrap_df['BAAA'] - bootstrap_df['ABAA']))\
        /((bootstrap_df['ABBA_HOM'] - bootstrap_df['BABA_HOM']) + (bootstrap_df['BAAA_HOM'] - bootstrap_df['ABAA_HOM']))
    # Extract information needed for hypothesis testing.
    # CEU.
    CEU_bootstrap_d = bootstrap_df[bootstrap_df['POP'] == 'CEU']['D'].values
    CEU_bootstrap_danc = bootstrap_df[bootstrap_df['POP'] == 'CEU']['Danc'].values
    CEU_bootstrap_dplus = bootstrap_df[bootstrap_df['POP'] == 'CEU']['D+'].values
    CEU_bootstrap_fhom = bootstrap_df[bootstrap_df['POP'] == 'CEU']['fhom'].values
    CEU_bootstrap_fanc = bootstrap_df[bootstrap_df['POP'] == 'CEU']['fanc'].values
    CEU_bootstrap_fplus = bootstrap_df[bootstrap_df['POP'] == 'CEU']['f+'].values
    # FIN.
    FIN_bootstrap_d = bootstrap_df[bootstrap_df['POP'] == 'FIN']['D'].values
    FIN_bootstrap_danc = bootstrap_df[bootstrap_df['POP'] == 'FIN']['Danc'].values
    FIN_bootstrap_dplus = bootstrap_df[bootstrap_df['POP'] == 'FIN']['D+'].values
    FIN_bootstrap_fhom = bootstrap_df[bootstrap_df['POP'] == 'FIN']['fhom'].values
    FIN_bootstrap_fanc = bootstrap_df[bootstrap_df['POP'] == 'FIN']['fanc'].values
    FIN_bootstrap_fplus = bootstrap_df[bootstrap_df['POP'] == 'FIN']['f+'].values
    # GBR.
    GBR_bootstrap_d = bootstrap_df[bootstrap_df['POP'] == 'GBR']['D'].values
    GBR_bootstrap_danc = bootstrap_df[bootstrap_df['POP'] == 'GBR']['Danc'].values
    GBR_bootstrap_dplus = bootstrap_df[bootstrap_df['POP'] == 'GBR']['D+'].values
    GBR_bootstrap_fhom = bootstrap_df[bootstrap_df['POP'] == 'GBR']['fhom'].values
    GBR_bootstrap_fanc = bootstrap_df[bootstrap_df['POP'] == 'GBR']['fanc'].values
    GBR_bootstrap_fplus = bootstrap_df[bootstrap_df['POP'] == 'GBR']['f+'].values
    # IBS.
    IBS_bootstrap_d = bootstrap_df[bootstrap_df['POP'] == 'IBS']['D'].values
    IBS_bootstrap_danc = bootstrap_df[bootstrap_df['POP'] == 'IBS']['Danc'].values
    IBS_bootstrap_dplus = bootstrap_df[bootstrap_df['POP'] == 'IBS']['D+'].values
    IBS_bootstrap_fhom = bootstrap_df[bootstrap_df['POP'] == 'IBS']['fhom'].values
    IBS_bootstrap_fanc = bootstrap_df[bootstrap_df['POP'] == 'IBS']['fanc'].values
    IBS_bootstrap_fplus = bootstrap_df[bootstrap_df['POP'] == 'IBS']['f+'].values
    # TSI.
    TSI_bootstrap_d = bootstrap_df[bootstrap_df['POP'] == 'TSI']['D'].values
    TSI_bootstrap_danc = bootstrap_df[bootstrap_df['POP'] == 'TSI']['Danc'].values
    TSI_bootstrap_dplus = bootstrap_df[bootstrap_df['POP'] == 'TSI']['D+'].values
    TSI_bootstrap_fhom = bootstrap_df[bootstrap_df['POP'] == 'TSI']['fhom'].values
    TSI_bootstrap_fanc = bootstrap_df[bootstrap_df['POP'] == 'TSI']['fanc'].values
    TSI_bootstrap_fplus = bootstrap_df[bootstrap_df['POP'] == 'TSI']['f+'].values
    # CHB.
    CHB_bootstrap_d = bootstrap_df[bootstrap_df['POP'] == 'CHB']['D'].values
    CHB_bootstrap_danc = bootstrap_df[bootstrap_df['POP'] == 'CHB']['Danc'].values
    CHB_bootstrap_dplus = bootstrap_df[bootstrap_df['POP'] == 'CHB']['D+'].values
    CHB_bootstrap_fhom = bootstrap_df[bootstrap_df['POP'] == 'CHB']['fhom'].values
    CHB_bootstrap_fanc = bootstrap_df[bootstrap_df['POP'] == 'CHB']['fanc'].values
    CHB_bootstrap_fplus = bootstrap_df[bootstrap_df['POP'] == 'CHB']['f+'].values
    # CHS.
    CHS_bootstrap_d = bootstrap_df[bootstrap_df['POP'] == 'CHS']['D'].values
    CHS_bootstrap_danc = bootstrap_df[bootstrap_df['POP'] == 'CHS']['Danc'].values
    CHS_bootstrap_dplus = bootstrap_df[bootstrap_df['POP'] == 'CHS']['D+'].values
    CHS_bootstrap_fhom = bootstrap_df[bootstrap_df['POP'] == 'CHS']['fhom'].values
    CHS_bootstrap_fanc = bootstrap_df[bootstrap_df['POP'] == 'CHS']['fanc'].values
    CHS_bootstrap_fplus = bootstrap_df[bootstrap_df['POP'] == 'CHS']['f+'].values
    # CDX.
    CDX_bootstrap_d = bootstrap_df[bootstrap_df['POP'] == 'CDX']['D'].values
    CDX_bootstrap_danc = bootstrap_df[bootstrap_df['POP'] == 'CDX']['Danc'].values
    CDX_bootstrap_dplus = bootstrap_df[bootstrap_df['POP'] == 'CDX']['D+'].values
    CDX_bootstrap_fhom = bootstrap_df[bootstrap_df['POP'] == 'CDX']['fhom'].values
    CDX_bootstrap_fanc = bootstrap_df[bootstrap_df['POP'] == 'CDX']['fanc'].values
    CDX_bootstrap_fplus = bootstrap_df[bootstrap_df['POP'] == 'CDX']['f+'].values
    # JPT.
    JPT_bootstrap_d = bootstrap_df[bootstrap_df['POP'] == 'JPT']['D'].values
    JPT_bootstrap_danc = bootstrap_df[bootstrap_df['POP'] == 'JPT']['Danc'].values
    JPT_bootstrap_dplus = bootstrap_df[bootstrap_df['POP'] == 'JPT']['D+'].values
    JPT_bootstrap_fhom = bootstrap_df[bootstrap_df['POP'] == 'JPT']['fhom'].values
    JPT_bootstrap_fanc = bootstrap_df[bootstrap_df['POP'] == 'JPT']['fanc'].values
    JPT_bootstrap_fplus = bootstrap_df[bootstrap_df['POP'] == 'JPT']['f+'].values
    # KHV.
    KHV_bootstrap_d = bootstrap_df[bootstrap_df['POP'] == 'KHV']['D'].values
    KHV_bootstrap_danc = bootstrap_df[bootstrap_df['POP'] == 'KHV']['Danc'].values
    KHV_bootstrap_dplus = bootstrap_df[bootstrap_df['POP'] == 'KHV']['D+'].values
    KHV_bootstrap_fhom = bootstrap_df[bootstrap_df['POP'] == 'KHV']['fhom'].values
    KHV_bootstrap_fanc = bootstrap_df[bootstrap_df['POP'] == 'KHV']['fanc'].values
    KHV_bootstrap_fplus = bootstrap_df[bootstrap_df['POP'] == 'KHV']['f+'].values
    # BEB.
    BEB_bootstrap_d = bootstrap_df[bootstrap_df['POP'] == 'BEB']['D'].values
    BEB_bootstrap_danc = bootstrap_df[bootstrap_df['POP'] == 'BEB']['Danc'].values
    BEB_bootstrap_dplus = bootstrap_df[bootstrap_df['POP'] == 'BEB']['D+'].values
    BEB_bootstrap_fhom = bootstrap_df[bootstrap_df['POP'] == 'BEB']['fhom'].values
    BEB_bootstrap_fanc = bootstrap_df[bootstrap_df['POP'] == 'BEB']['fanc'].values
    BEB_bootstrap_fplus = bootstrap_df[bootstrap_df['POP'] == 'BEB']['f+'].values
    # GIH.
    GIH_bootstrap_d = bootstrap_df[bootstrap_df['POP'] == 'GIH']['D'].values
    GIH_bootstrap_danc = bootstrap_df[bootstrap_df['POP'] == 'GIH']['Danc'].values
    GIH_bootstrap_dplus = bootstrap_df[bootstrap_df['POP'] == 'GIH']['D+'].values
    GIH_bootstrap_fhom = bootstrap_df[bootstrap_df['POP'] == 'GIH']['fhom'].values
    GIH_bootstrap_fanc = bootstrap_df[bootstrap_df['POP'] == 'GIH']['fanc'].values
    GIH_bootstrap_fplus = bootstrap_df[bootstrap_df['POP'] == 'GIH']['f+'].values
    # ITU.
    ITU_bootstrap_d = bootstrap_df[bootstrap_df['POP'] == 'ITU']['D'].values
    ITU_bootstrap_danc = bootstrap_df[bootstrap_df['POP'] == 'ITU']['Danc'].values
    ITU_bootstrap_dplus = bootstrap_df[bootstrap_df['POP'] == 'ITU']['D+'].values
    ITU_bootstrap_fhom = bootstrap_df[bootstrap_df['POP'] == 'ITU']['fhom'].values
    ITU_bootstrap_fanc = bootstrap_df[bootstrap_df['POP'] == 'ITU']['fanc'].values
    ITU_bootstrap_fplus = bootstrap_df[bootstrap_df['POP'] == 'ITU']['f+'].values
    # PJL.
    PJL_bootstrap_d = bootstrap_df[bootstrap_df['POP'] == 'PJL']['D'].values
    PJL_bootstrap_danc = bootstrap_df[bootstrap_df['POP'] == 'PJL']['Danc'].values
    PJL_bootstrap_dplus = bootstrap_df[bootstrap_df['POP'] == 'PJL']['D+'].values
    PJL_bootstrap_fhom = bootstrap_df[bootstrap_df['POP'] == 'PJL']['fhom'].values
    PJL_bootstrap_fanc = bootstrap_df[bootstrap_df['POP'] == 'PJL']['fanc'].values
    PJL_bootstrap_fplus = bootstrap_df[bootstrap_df['POP'] == 'PJL']['f+'].values
    # STU.
    STU_bootstrap_d = bootstrap_df[bootstrap_df['POP'] == 'STU']['D'].values
    STU_bootstrap_danc = bootstrap_df[bootstrap_df['POP'] == 'STU']['Danc'].values
    STU_bootstrap_dplus = bootstrap_df[bootstrap_df['POP'] == 'STU']['D+'].values
    STU_bootstrap_fhom = bootstrap_df[bootstrap_df['POP'] == 'STU']['fhom'].values
    STU_bootstrap_fanc = bootstrap_df[bootstrap_df['POP'] == 'STU']['fanc'].values
    STU_bootstrap_fplus = bootstrap_df[bootstrap_df['POP'] == 'STU']['f+'].values
    # CLM.
    CLM_bootstrap_d = bootstrap_df[bootstrap_df['POP'] == 'CLM']['D'].values
    CLM_bootstrap_danc = bootstrap_df[bootstrap_df['POP'] == 'CLM']['Danc'].values
    CLM_bootstrap_dplus = bootstrap_df[bootstrap_df['POP'] == 'CLM']['D+'].values
    CLM_bootstrap_fhom = bootstrap_df[bootstrap_df['POP'] == 'CLM']['fhom'].values
    CLM_bootstrap_fanc = bootstrap_df[bootstrap_df['POP'] == 'CLM']['fanc'].values
    CLM_bootstrap_fplus = bootstrap_df[bootstrap_df['POP'] == 'CLM']['f+'].values
    # MXL.
    MXL_bootstrap_d = bootstrap_df[bootstrap_df['POP'] == 'MXL']['D'].values
    MXL_bootstrap_danc = bootstrap_df[bootstrap_df['POP'] == 'MXL']['Danc'].values
    MXL_bootstrap_dplus = bootstrap_df[bootstrap_df['POP'] == 'MXL']['D+'].values
    MXL_bootstrap_fhom = bootstrap_df[bootstrap_df['POP'] == 'MXL']['fhom'].values
    MXL_bootstrap_fanc = bootstrap_df[bootstrap_df['POP'] == 'MXL']['fanc'].values
    MXL_bootstrap_fplus = bootstrap_df[bootstrap_df['POP'] == 'MXL']['f+'].values
    # PEL.
    PEL_bootstrap_d = bootstrap_df[bootstrap_df['POP'] == 'PEL']['D'].values
    PEL_bootstrap_danc = bootstrap_df[bootstrap_df['POP'] == 'PEL']['Danc'].values
    PEL_bootstrap_dplus = bootstrap_df[bootstrap_df['POP'] == 'PEL']['D+'].values
    PEL_bootstrap_fhom = bootstrap_df[bootstrap_df['POP'] == 'PEL']['fhom'].values
    PEL_bootstrap_fanc = bootstrap_df[bootstrap_df['POP'] == 'PEL']['fanc'].values
    PEL_bootstrap_fplus = bootstrap_df[bootstrap_df['POP'] == 'PEL']['f+'].values
    # PUR.
    PUR_bootstrap_d = bootstrap_df[bootstrap_df['POP'] == 'PUR']['D'].values
    PUR_bootstrap_danc = bootstrap_df[bootstrap_df['POP'] == 'PUR']['Danc'].values
    PUR_bootstrap_dplus = bootstrap_df[bootstrap_df['POP'] == 'PUR']['D+'].values
    PUR_bootstrap_fhom = bootstrap_df[bootstrap_df['POP'] == 'PUR']['fhom'].values
    PUR_bootstrap_fanc = bootstrap_df[bootstrap_df['POP'] == 'PUR']['fanc'].values
    PUR_bootstrap_fplus = bootstrap_df[bootstrap_df['POP'] == 'PUR']['f+'].values
    # Create introgression metrics standard deviation arrays.
    bootstrap_d_stds = np.array([
        np.std(CEU_bootstrap_d), np.std(FIN_bootstrap_d), np.std(GBR_bootstrap_d), np.std(IBS_bootstrap_d), np.std(TSI_bootstrap_d),
        np.std(CHB_bootstrap_d), np.std(CHS_bootstrap_d), np.std(CDX_bootstrap_d), np.std(JPT_bootstrap_d), np.std(KHV_bootstrap_d),
        np.std(BEB_bootstrap_d), np.std(GIH_bootstrap_d), np.std(ITU_bootstrap_d), np.std(PJL_bootstrap_d), np.std(STU_bootstrap_d),
        np.std(CLM_bootstrap_d), np.std(MXL_bootstrap_d), np.std(PEL_bootstrap_d), np.std(PUR_bootstrap_d),
    ])
    bootstrap_danc_stds = np.array([
        np.std(CEU_bootstrap_danc), np.std(FIN_bootstrap_danc), np.std(GBR_bootstrap_danc), np.std(IBS_bootstrap_danc), np.std(TSI_bootstrap_danc),
        np.std(CHB_bootstrap_danc), np.std(CHS_bootstrap_danc), np.std(CDX_bootstrap_danc), np.std(JPT_bootstrap_danc), np.std(KHV_bootstrap_danc),
        np.std(BEB_bootstrap_danc), np.std(GIH_bootstrap_danc), np.std(ITU_bootstrap_danc), np.std(PJL_bootstrap_danc), np.std(STU_bootstrap_danc),
        np.std(CLM_bootstrap_danc), np.std(MXL_bootstrap_danc), np.std(PEL_bootstrap_danc), np.std(PUR_bootstrap_danc),
    ])
    bootstrap_dplus_stds = np.array([
        np.std(CEU_bootstrap_dplus), np.std(FIN_bootstrap_dplus), np.std(GBR_bootstrap_dplus), np.std(IBS_bootstrap_dplus), np.std(TSI_bootstrap_dplus),
        np.std(CHB_bootstrap_dplus), np.std(CHS_bootstrap_dplus), np.std(CDX_bootstrap_dplus), np.std(JPT_bootstrap_dplus), np.std(KHV_bootstrap_dplus),
        np.std(BEB_bootstrap_dplus), np.std(GIH_bootstrap_dplus), np.std(ITU_bootstrap_dplus), np.std(PJL_bootstrap_dplus), np.std(STU_bootstrap_dplus),
        np.std(CLM_bootstrap_dplus), np.std(MXL_bootstrap_dplus), np.std(PEL_bootstrap_dplus), np.std(PUR_bootstrap_dplus),
    ])
    bootstrap_fhom_stds = np.array([
        np.std(CEU_bootstrap_fhom), np.std(FIN_bootstrap_fhom), np.std(GBR_bootstrap_fhom), np.std(IBS_bootstrap_fhom), np.std(TSI_bootstrap_fhom),
        np.std(CHB_bootstrap_fhom), np.std(CHS_bootstrap_fhom), np.std(CDX_bootstrap_fhom), np.std(JPT_bootstrap_fhom), np.std(KHV_bootstrap_fhom),
        np.std(BEB_bootstrap_fhom), np.std(GIH_bootstrap_fhom), np.std(ITU_bootstrap_fhom), np.std(PJL_bootstrap_fhom), np.std(STU_bootstrap_fhom),
        np.std(CLM_bootstrap_fhom), np.std(MXL_bootstrap_fhom), np.std(PEL_bootstrap_fhom), np.std(PUR_bootstrap_fhom),
    ])
    bootstrap_fanc_stds = np.array([
        np.std(CEU_bootstrap_fanc), np.std(FIN_bootstrap_fanc), np.std(GBR_bootstrap_fanc), np.std(IBS_bootstrap_fanc), np.std(TSI_bootstrap_fanc),
        np.std(CHB_bootstrap_fanc), np.std(CHS_bootstrap_fanc), np.std(CDX_bootstrap_fanc), np.std(JPT_bootstrap_fanc), np.std(KHV_bootstrap_fanc),
        np.std(BEB_bootstrap_fanc), np.std(GIH_bootstrap_fanc), np.std(ITU_bootstrap_fanc), np.std(PJL_bootstrap_fanc), np.std(STU_bootstrap_fanc),
        np.std(CLM_bootstrap_fanc), np.std(MXL_bootstrap_fanc), np.std(PEL_bootstrap_fanc), np.std(PUR_bootstrap_fanc),
    ])
    bootstrap_fplus_stds = np.array([
        np.std(CEU_bootstrap_fplus), np.std(FIN_bootstrap_fplus), np.std(GBR_bootstrap_fplus), np.std(IBS_bootstrap_fplus), np.std(TSI_bootstrap_fplus),
        np.std(CHB_bootstrap_fplus), np.std(CHS_bootstrap_fplus), np.std(CDX_bootstrap_fplus), np.std(JPT_bootstrap_fplus), np.std(KHV_bootstrap_fplus),
        np.std(BEB_bootstrap_fplus), np.std(GIH_bootstrap_fplus), np.std(ITU_bootstrap_fplus), np.std(PJL_bootstrap_fplus), np.std(STU_bootstrap_fplus),
        np.std(CLM_bootstrap_fplus), np.std(MXL_bootstrap_fplus), np.std(PEL_bootstrap_fplus), np.std(PUR_bootstrap_fplus),
    ])
    # Load the population level data frame.
    genome_df = load_population_results()
    # Calculate p-values.
    genome_df['D_bs_std'] = bootstrap_d_stds
    genome_df['Danc_bs_std'] = bootstrap_danc_stds
    genome_df['Dplus_bs_std'] = bootstrap_dplus_stds
    genome_df['D_pval'] = norm.sf(x=abs(genome_df['D']), loc=0, scale=genome_df['D_bs_std'])
    genome_df['Danc_pval'] = norm.sf(x=abs(genome_df['Danc']), loc=0, scale=genome_df['Danc_bs_std'])
    genome_df['Dplus_pval'] = norm.sf(x=abs(genome_df['D+']), loc=0, scale=genome_df['Dplus_bs_std'])
    # Intialize the population order.
    tgp_pop_array = np.array([
        'CEU', 'FIN', 'GBR', 'IBS', 'TSI',
        'CHB', 'CHS', 'CDX', 'JPT', 'KHV',
        'BEB', 'GIH', 'ITU', 'PJL', 'STU',
        'CLM', 'MXL', 'PEL', 'PUR',
    ])
    # Construct the final data frame.
    results_df = pd.DataFrame(
        data={
            r'$ABBA$': np.round(genome_df['ABBA'].values, decimals=3),
            r'$BABA$': np.round(genome_df['BABA'].values, decimals=3),
            r'$BAAA$': np.round(genome_df['BAAA'].values, decimals=3),
            r'$ABAA$': np.round(genome_df['ABAA'].values, decimals=3),
            r'$ABBA-BABA$': np.round(genome_df['ABBA-BABA'].values, decimals=3),
            r'$BAAA-ABAA$': np.round(genome_df['BAAA-ABAA'].values, decimals=3),
            r'$D$': np.round(genome_df['D'].values, decimals=3),
            r'$D \;(p-value)$': genome_df['D_pval'].values,
            r'$D_{anc}$': np.round(genome_df['Danc'].values, decimals=3),
            r'$D_{anc} \;(p-value)$': genome_df['Danc_pval'].values,
            r'$D+$': np.round(genome_df['D+'].values, decimals=3),
            r'$D+ \;(p-value)$': genome_df['Dplus_pval'].values,
            r'$f_{hom}$': np.round(genome_df['fhom'].values, decimals=3),
            r'$f_{hom} \;(\sigma)$': np.round(bootstrap_fhom_stds, decimals=3),
            r'$f_{anc}$': np.round(genome_df['fanc'].values, decimals=3),
            r'$f_{anc} \;(\sigma)$': np.round(bootstrap_fanc_stds, decimals=3),
            r'$f+$': np.round(genome_df['f+'].values, decimals=3),
            r'$f+ \;(\sigma)$': np.round(bootstrap_fplus_stds, decimals=3),
        },
        index=tgp_pop_array,
    )
    return results_df
```

```python
# LaTeX plotting helper function provided by Jack Walton (https://jwalton.info/Embed-Publication-Matplotlib-Latex/).
def set_size(width_pt=452, fraction=1, subplots=(1, 1)):
    """Set figure dimensions to avoid scaling in LaTeX.

    Parameters
    ----------
    width: float or string
            Document width in points, or string of predined document type
    fraction: float, optional
            Fraction of the width which you wish the figure to occupy
    subplots: array-like, optional
            The number of rows and columns of subplots.
    Returns
    -------
    fig_dim: tuple
            Dimensions of figure in inches
    """
    # Width of figure (in pts)
    fig_width_pt = width_pt * fraction
    # Convert from pt to inches
    inches_per_pt = 1 / 72.27
    # Golden ratio to set aesthetic figure height.
    golden_ratio = (5**.5 - 1) / 2
    # Figure width in inches.
    fig_width_in = fig_width_pt * inches_per_pt
    # Figure height in inches.
    fig_height_in = fig_width_in * golden_ratio * (subplots[0] / subplots[1])
    return (fig_width_in, fig_height_in)
# Define the max dimesnions for the manuscript.
max_dim = (6.25432406254324, 3.865384847308046)
```

```python
# Intialize the TGP populations.
tgp_trio_pops = [
    'CEU', 'FIN', 'GBR', 'IBS', 'TSI',
    'CHB', 'CHS', 'CDX', 'JPT', 'KHV',
    'BEB', 'GIH', 'ITU', 'PJL', 'STU',
    'CLM', 'MXL', 'PEL', 'PUR',
]
# Intialize an empty dictionary.
trios_genome_dicc = {}
# For every population...
for pop in tgp_trio_pops:
    # Intialize the site pattern totals.
    trios_genome_dicc[pop] = {}
    abba = np.zeros(trios_all[1][pop]['ABBA'].size)
    baba = np.zeros(trios_all[1][pop]['BABA'].size)
    baaa = np.zeros(trios_all[1][pop]['BAAA'].size)
    abaa = np.zeros(trios_all[1][pop]['ABAA'].size)
    # For every chromosome...
    for chrom in range(1, 23):
        # Update the site pattern counts.
        abba += trios_all[chrom][pop]['ABBA']
        baba += trios_all[chrom][pop]['BABA']
        baaa += trios_all[chrom][pop]['BAAA']
        abaa += trios_all[chrom][pop]['ABAA']
    trios_genome_dicc[pop]['ABBA'] = abba
    trios_genome_dicc[pop]['BABA'] = baba
    trios_genome_dicc[pop]['ABBA-BABA'] = (abba - baba)
    trios_genome_dicc[pop]['BAAA'] = baaa
    trios_genome_dicc[pop]['ABAA'] = abaa
    trios_genome_dicc[pop]['BAAA-ABAA'] = (baaa - abaa)
    trios_genome_dicc[pop]['DIFF'] = (abba - baba) - (baaa - abaa)
# Intialize a dataframe dictionary.
diff_df_dicc = {
    r'$ABBA-BABA$': [],
    r'$BAAA-ABAA$': [],
    r'$P-value$': [],
}
# For every population...
for key in trios_genome_dicc.keys():
    # Fill the dictionary.
    mean = np.mean(trios_genome_dicc[key]['DIFF'])
    std = np.std(trios_genome_dicc[key]['DIFF'])
    pval = norm.sf(x=abs(mean), loc=0, scale=std)
    diff_df_dicc[r'$ABBA-BABA$'].append(np.mean(trios_genome_dicc[key]['ABBA-BABA']))
    diff_df_dicc[r'$BAAA-ABAA$'].append(np.mean(trios_genome_dicc[key]['BAAA-ABAA']))
    diff_df_dicc[r'$P-value$'].append(pval)
# Construct the results data frame.
trios_diff_df = pd.DataFrame(
    data=diff_df_dicc,
    index=tgp_index_array,
)
trios_diff_df
```

```python
# Intialize the table header.
header = r'\begin{tabular}{@{}lccc@{}}'+'\n'\
+r'\toprule'+'\n'\
+r' & $ABBA-BABA$ & $BAAA-ABAA$ & $p-value$ \\'+'\n'\
+r'\midrule'+'\n'\
# Intialize the table footer.
footer = r'\bottomrule'+'\n'+r'\end{tabular}'+'\n'
# Intialize a datframe matrix.
df_mat = trios_diff_df.to_numpy()
# Open the table.
table = open('../../manuscript/tables/tgp_altai_trios_site_pattern_diff.tex', 'w')
# Write the header.
table.write(header)
# For every admixture proportion.
for idx in range(tgp_index_array.size):
    # Write the results.
    table.write(
        r'{0} & {1} & {2} & {3} \\'.format(
            tgp_index_array[idx],
            round(df_mat[idx, 0], 3), round(df_mat[idx, 1], 3), round(df_mat[idx, 2], 3),
        )+'\n')
# Write the footer.
table.write(footer)
# Close the table file.
table.close()
```
